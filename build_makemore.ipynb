```python
words = open("cities.txt", "r").read().splitlines()
```

```python
words
```

```python
min(len(w) for w in words)
```

```python
max(len(w) for w in words)
```

```python
b = {}
for l in words:
    chs = ["<S>"] + list(l) + ["<E>"]
    for ch1, ch2 in zip(chs, chs[1:]):
        bigram = (ch1, ch2)
        b[bigram] = b.get(bigram, 0) + 1
```
